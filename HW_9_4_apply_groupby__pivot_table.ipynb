## 2. Розподіл кількості покупок за статтю. Які користувачі - жінки чи чоловіки більше купують?

Розглянемо розподілення покупок за статтю в різних містах:

```python
 df.groupby(['City', 'Gender']).agg(Invoces = ('Invoice ID', 'count'), Total = ('Total','sum'), Mean = ('Total','mean'))
```

Варто розробляти маркетингові стратегії, орієнтовані більш на жінок, адже вони формують більшу частку виручки.
- Mandalay: Кількість і суми для чоловіків і жінок приблизно рівні, середній чек у жінок трохи вищий, але різниця не суттєва.
- Yangon: Хоча чоловіків більше за кількістю інвойсів, середній чек у них нижчий за жіночий, що компенсує загальну суму продажів і робить її приблизно рівною між статями. В Yangon має сенс підвищувати середній чек серед чоловіків за допомогою акцій чи пакетних пропозицій, бо активність у них є, але покупки поки що менш коштовні.

## 3. Сезонність або часові тренди.
Чи змінюються продажі у різні періоди року в містах? Наприклад, чи є сезонність по місяцях, та збільшення продажів по вихідних днях.

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
df.loc[1, 'Date']
```

```python
df['YearMonth'] = df['Date'].dt.to_period('M')
df['Weekday'] = df['Date'].dt.day_name()
df.head(5)
```

```python
# По місяцях

df.groupby(['City', 'YearMonth']).agg(Invoces = ('Invoice ID', 'count'), Total = ('Total','sum'))
```

**Висновок:**
*   Даних замало для аналізу сезонності, але бачимо, що є загальний тренд для всіх міст. В січні та березні збільшується і кількість інвойсів, і загальний прибуток у всіх містах.


Проаналізуємо прожажі за днями тижня.

```python
# Кількість інвойсів за днями тижня
sales_by_day = df['Weekday'].value_counts()
print(sales_by_day)
```

```python
df_Mandalay.groupby('Weekday').agg(Invoces = ('Invoice ID', 'count'), Total = ('Total','sum')).sort_values(by='Total', ascending=False)
```

**Висновок:**
Для міста Mandalay найприбутковішими днями тижня виявилися **субота та вівторок**. А неділя виявилася найменш прибутковою, це може бути пов'язано з режимом роботи, можливо в неділю машазин працює не повний день.

```python
df_Naypyitaw.groupby('Weekday').agg(Invoces = ('Invoice ID', 'count'), Total = ('Total','sum')).sort_values(by='Total', ascending=False)
```

**Висновок:**

```python
df_Yangon = df[df['City'] == 'Yangon']
```

```python
df_Yangon.groupby('Weekday').agg(Invoces = ('Invoice ID', 'count'), Total = ('Total','sum')).sort_values(by='Total', ascending=False)
```

**Висновок:**
Для міста Yangon найприбутковішими днями тижня виявилися **неділя** та **субота**. А середа виявилася найменш прибутковою.

**Висновок**:
Спостерігається загальний тренд для всіх міст, що в суботу магазини мають найбільшу кількість інвойсів та прибутку. Це пов'язано з вихідним днем, люди мають час зробити покупки та прогулятися магазинами.
Щодо неділі, то тільки в місті Mandalay низький прибуток. Можливо там магазин працює неповний день, і треба змінити графік роботи, бо інщі філіали показують високий прибуток в неділю.
